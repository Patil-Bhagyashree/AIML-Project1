```python
class SimpleChatbot:
    def __init__(self):
        self.user_name = ""
        self.likes_pizza = None
        self.favorite_color = ""
        self.previous_questions_asked = []
        self.responses_to_unknown = ["Sorry, I didn't get that.", "Can you say that differently?", "I'm not sure I understand."]

    def greet_user(self):
        print("Hello! I am a simple chatbot. What's your name?")
        self.user_name = input("> ")
        print(f"Nice to meet you, {self.user_name}!")

    def ask_questions(self):
        questions = [
            "Do you like pizza? (yes/no)",
            "What's your favorite color?"
        ]
        for question in questions:
            if question not in self.previous_questions_asked:
                print(question)
                response = input("> ").lower().strip()
                self.handle_response(question, response)
                self.previous_questions_asked.append(question)
                break  # Break after asking one question to simulate conversation flow

    def handle_response(self, question, response):
        if question == "Do you like pizza? (yes/no)":
            if response in ["yes", "y"]:
                self.likes_pizza = True
                print("Pizza is great!")
            elif response in ["no", "n"]:
                self.likes_pizza = False
                print("More pizza for me then!")
            else:
                print("Just a simple yes or no will do.")
        elif question == "What's your favorite color?":
            self.favorite_color = response
            print(f"{self.favorite_color} is a nice color!")

    def recall_previous_interaction(self):
        if self.user_name:
            print(f"By the way, {self.user_name}, I remember you mentioned your favorite color was {self.favorite_color}.")
            if self.likes_pizza is not None:
                pizza_preference = "like" if self.likes_pizza else "don't like"
                print(f"And you said you {pizza_preference} pizza!")

    def respond_to_basic_questions(self, question):
        basic_answers = {
            "how are you?": "I'm a chatbot, so always good!",
            "what is your name?": "I'm a simple chatbot.",
            "what do you do?": "I chat with people!",
            "where do you live?": "In the cloud.",
            "do you have feelings?": "No, I'm not programmed for that."
        }
        return basic_answers.get(question.lower().strip(), None)

    def say_farewell(self):
        print("It was nice talking to you. Goodbye!")

    def chat(self):
        self.greet_user()
        chatting = True
        while chatting:
            print("Ask me a question, or type 'quit' to leave.")
            user_input = input("> ").lower().strip()
            if user_input == "quit":
                self.say_farewell()
                chatting = False
            elif user_input in ["recall", "remember"]:
                self.recall_previous_interaction()
            else:
                answer = self.respond_to_basic_questions(user_input)
                if answer:
                    print(answer)
                else:
                    print(self.responses_to_unknown[0])  # Simple error handling
                    self.ask_questions()  # Ask the user a question if we don't understand theirs

if __name__ == "__main__":
    chatbot = SimpleChatbot()
    chatbot.chat()
```
